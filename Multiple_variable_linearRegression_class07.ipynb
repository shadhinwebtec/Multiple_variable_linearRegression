# **Linear Regression with Gradient Desent**

```python
m = 0
c = 0
learning_rate = 0.0001

n = int(len(x))

iterations = 1000

for i in range(iterations):
    pred_y = m*x + c
    der_m = (-2/n) * (sum(x * (y-pred_y)))
    der_c = (-2/n) * (sum(y - pred_y))
    
    m = m - (learning_rate * der_m)
    c = c - (learning_rate * der_c)
    print(m,c)
```

```python
x = float(input('Enter any x:'))
y = (m * x ) + c
print(y)
```
